First generating fig 2.2 i.e. e-greedy method for stationary for 10 armed bandit.
We will do the following experiment for epsion = 0, 0.01 and 0.1:

    We will do the following 2000 times:

    1. a = 1,2,3,...,10 is the set of actions. 
    2. Initialise Q(a)=0 for each action a.
    3. Choose the value for each action randomly from gaussian(mean=0,var=1) and let it be q*(a) for each a.
    4. Now do following for 1000 steps (using e-greedy):
        i) let action At be chosen.
        ii) Choose the reward Rt from gaussian(mean=q*(a),var=1)
        iii) Change the estimate for Q(At)

Calculate average reward and plot it against no of steps. Do this for all three epsilons in one graph.
Plot optimal action (percentage) vs steps for each epsilon.

```python
%matplotlib inline
import numpy as np
import math
from numpy import random as rand
import matplotlib.pyplot as plt

# function to return an action based on epsilon greedy approach
def getAction(q_action,epsilon):
    if(rand.random()<epsilon):
        return rand.randint(0,len(q_action))
    else:
        mx=-1*math.inf
        for each in q_action:
            if(each>mx):
                mx=each
        mx_indices = []
        i=0
        for each in q_action:
            if(each==mx):
                mx_indices.append(i)
            i+=1
        return mx_indices[rand.randint(0,len(mx_indices))]
```

```python
def exp1():
    print("starting exp1")
    plt.figure(figsize=(18,16))
    g1 = plt.subplot(2,1,1)
    plt.xlabel('Steps') 
    plt.ylabel('Average Reward') 
    g2 = plt.subplot(2,1,2)
    plt.xlabel('Steps')
    plt.ylabel('% Optimal action')
    for epsilon in [0,0.01,0.1]:
        rewards=[] #average reward at each step
        steps=[] #step number
        optimal=[] #at each step, avg chance of choosing the optimal action
        for i in range(0,1000):
            rewards.append(0)
            steps.append(i+1)
            optimal.append(0)
        for exp in range(2000):
            Q_action=[] #actual
            q_action=[] #estimated
            n_action=[] #no. of times this action was taken
            opt_action = []
            opt_value = -1*math.inf
            for i in range(10):
                Q_action.append(rand.normal(0,1))
                q_action.append(0)
                n_action.append(0)
                if(Q_action[i]>opt_value):
                    opt_value=Q_action[i]
            for i in range(10):
                if(Q_action[i]==opt_value):
                    opt_action.append(i)
            for i in range(1000):
                action = getAction(q_action,epsilon)
                if(action in opt_action):
                    optimal[i]+=1
                n_action[action]+=1
                reward = rand.normal(Q_action[action],1)
                rewards[i]+=reward
                q_action[action] = q_action[action]+(reward-q_action[action])/(n_action[action])
        for i in range(0,1000):
            rewards[i]/=2000
            optimal[i]/=20
        print("done for epsilon = "+str(epsilon))
        g1.plot(steps,rewards,label=str(epsilon))
        g2.plot(steps,optimal,label=str(epsilon))
    g1.legend()
    g2.legend()
    plt.show()
    print("done exp1")
    
exp1()
    
            
        
```

**Q1 - Exercice 2.5**

Exercice 2.5 An experiment to demonstrate the difficulties that sample-average methods have for nonstationary problems
    10 armed testbed with epsilon=0.1:
    1. All q*(a) start out equal 
    2. Increment all q*(a) with a random value from gauss(0,0.01) at each step. (Generate new randoms for each action at each step).
    3. Plot like above for the two algorithms and execute 10000 steps:
        i) using sample average method 
        ii) using constant step size method with alpha=0.1

```python
def exp2():
    num_steps=10000
    print("starting exp2")
    alpha=0.1
    plt.figure(figsize=(18,16))
    g1 = plt.subplot(2,1,1)
    plt.xlabel('Steps') 
    plt.ylabel('Average Reward') 
    g2 = plt.subplot(2,1,2)
    plt.xlabel('Steps')
    plt.ylabel('% Optimal action')
    for method in ["sample average","const step size"]:
        for epsilon in [0.1]:
            rewards=[] #average reward at each step
            steps=[] #step number
            optimal=[] #at each step, avg chance of choosing the optimal action
            for i in range(0,num_steps):
                rewards.append(0)
                steps.append(i+1)
                optimal.append(0)
            for exp in range(2000):
                Q_action=[] #actual
                q_action=[] #estimated
                n_action=[] #no. of times this action was taken
                ran_val=rand.normal(0,1)
                for i in range(10):
                    Q_action.append(ran_val) #as all start equal
                    q_action.append(0)
                    n_action.append(0)
                for i in range(num_steps):
                    ### calculating optimal
                    opt_action = []
                    opt_value = -1*math.inf
                    for j in range(10):
                        if(Q_action[j]>opt_value):
                            opt_value=Q_action[j]
                    for j in range(10):
                        if(Q_action[j]==opt_value):
                            opt_action.append(j)
                    ### -------------------
                    action = getAction(q_action,epsilon)
                    if(action in opt_action):
                        optimal[i]+=1
                    n_action[action]+=1
                    reward = rand.normal(Q_action[action],1)
                    rewards[i]+=reward
                    if(method=="sample average"):
                        q_action[action] = q_action[action]+(reward-q_action[action])/(n_action[action])
                    elif(method=="const step size"):
                        q_action[action] = q_action[action]+(reward-q_action[action])*(alpha)
                    for j in range(10):
                        Q_action[j]+=rand.normal(0,0.01)
            for i in range(0,num_steps):
                rewards[i]/=2000
                optimal[i]/=20
            print("done "+method+" for epsilon = "+str(epsilon))
            g1.plot(steps,rewards,label=method+" with epsilon="+str(epsilon))
            g2.plot(steps,optimal,label=method+" with epsilon="+str(epsilon))
    g1.legend()
    g2.legend()
    plt.show()
    print("done exp2")
    
exp2()
```

Explanation of above plot:
    We see that sample average performs worse than constant step size. This is because sample average gives equal weights to all rewards but constant step size gives more weight to more recent rewards to make estimates. Hence it is able to better estimate with the changing environment.

Answer to Exercise 2.6
In Fig 2.3, we see sudden oscillations/spikes in the early part of the curve because of the following reason:
At first, all values are highly optimistic i.e. 5 so each time we choose any action, we will definitely decrease the value. So in first 10 steps, we choose all actions once because all action values must have decreased by some amount on each step so that the other ones (which aren't chosen yet) become greedy choice till all are less than 5. Now, the action which had highest mean reward, its value would have decreased the least (on average) i.e. after first 10 steps, the action which had highest mean reward will have its value as the largest on average. So we will choose it again until it becomes less than all other values which will happen very fast as its estimate is still quite more (near 5) than what its actual value is. This is the reason we have spikes/oscillations in the start after around 10 steps because on average we will choose the most optimal action and thus model will perform better. Then, as and when it choses this action more, it decreases its estimate further, giving other actions a chance to be taken i.e exploration. Another thing to notice is that our curve start at 10 percent which is as expected because in the first 10 steps, we will choose each action once so probability of choosing optimal is 10 percent as there are 10 actions.

**Q4. Compare UCB to Optimistic Value and e-greedy for stationary case:**

```python
def ucb(q_action,c,t,n):
    res = []
    for i in range(0,len(q_action)):
        if(n[i]==0):
            res.append(math.inf)
        else:
            res.append(q_action[i]+c*math.sqrt(math.log(t)/n[i]))
    mx=-1*math.inf
    for each in res:
        if(each>mx):
            mx=each
    mx_indices = []
    i=0
    for each in res:
        if(each==mx):
            mx_indices.append(i)
        i+=1
    return mx_indices[rand.randint(0,len(mx_indices))]

def exp5():
    num_steps=5000
    print("starting exp5")
    alpha=0.1
    plt.figure(figsize=(18,8))
    plt.xlabel('Steps')
    plt.ylabel('% Optimal action')
    for algo in ["ucb","ov","eg"]:
        for method in ["sample average"]:
            for epsilon in [0,0.1]:
                if(not((algo=="ov" and epsilon==0) or (algo=="eg" and epsilon==0.1) or (algo=="ucb" and epsilon==0))):
                    continue
                rewards=[] #average reward at each step
                steps=[] #step number
                optimal=[] #at each step, avg chance of choosing the optimal action
                for i in range(0,num_steps):
                    rewards.append(0)
                    steps.append(i+1)
                    optimal.append(0)
                for exp in range(2000):
                    Q_action=[] #actual
                    q_action=[] #estimated
                    n_action=[] #no. of times this action was taken
                    opt_action = []
                    opt_value = -1*math.inf
                    for i in range(10):
                        Q_action.append(rand.normal(0,1))
                        if(algo=="ov"):
                            q_action.append(5)
                        else:
                            q_action.append(0)
                        n_action.append(0)
                        if(Q_action[i]>opt_value):
                            opt_value=Q_action[i]
                    for i in range(10):
                        if(Q_action[i]==opt_value):
                            opt_action.append(i)
                    for i in range(num_steps):
                        action=0
                        if(algo=="ucb"):
                            action = ucb(q_action,2,i+1,n_action)
                        else:
                            action = getAction(q_action,epsilon)
                        if(action in opt_action):
                            optimal[i]+=1
                        n_action[action]+=1
                        reward = rand.normal(Q_action[action],1)
                        rewards[i]+=reward
                        if(method=="sample average"):
                            q_action[action] = q_action[action]+(reward-q_action[action])/(n_action[action])
                        elif(method=="const step size"):
                            q_action[action] = q_action[action]+(reward-q_action[action])*(alpha)
                for i in range(0,num_steps):
                    rewards[i]/=2000
                    optimal[i]/=20
                print("done "+algo+" and "+method+" for epsilon = "+str(epsilon))
                plt.plot(steps,optimal,label=algo+" and "+method+" with epsilon="+str(epsilon))
    plt.legend()
    plt.show()
    print("done exp5")
    
exp5()
```

Explanation of above graph:
    Above case is of stationary model. We see that optimistic value performs worst followed by e-greedy and ucb performs the best. The technique chosen for averaging is sample mean because it gives equal weightage to all rewards and we are in a stationary setting so we would want that. Now we see that optimistic value method rises very quickly but over time it stays constant because it does not do much exploration. It is more or less greedy and since it wouldn't have learned q*(a), so it is biased based on its estimate of q(a) which is good 60 percent of the time. Now since e-greedy explores more often, so it is better than optimal in long run as it has a better estimate of q(a). UCB performs the best because c=2 encourages exploration and N(a) term decreases exploration for that action if it has been chosen enough so it kindof balances and gives the best of both worlds and optimises best.

Instead recent rewards should be given more weight. Now optimistic value performance worse because it basically bias its model according to the initial assumptions and doesnt explore much. E-greedy succeeds because it naturally explores as well so it changes its estimates according to the environment. But ucb falls somewhere in the middle because later in time steps, all actions have been chosen large enough so its exploration deacreases as log(t) terms doesn't increase that fast as compared to the denominator n(a) term.
